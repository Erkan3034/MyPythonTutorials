1'den 100'e kadar olan sayılardan pisagor üçgeni oluşturanları ekrana yazdıran bir fonksiyon yazalım.(a <= 100,b <= 100)

```python
import math

def pisagor_bulma():
    pisagor_listesi = []  # Boş bir liste oluşturalım.
    
    for i in range(1, 101):  # 1'den 100'e kadar olan sayılar
        for j in range(1, 101):
            c = math.sqrt(i ** 2 + j ** 2)  # Pisagor teoremi: c = sqrt(i^2 + j^2)
            
            if c == int(c):  # Eğer c tam sayıya eşitse ( 16 + 9 = 25)
                pisagor_listesi.append((i, j, int(c)))  # (i, j, c) üçlüsü listeye ekleyelm
                
    return pisagor_listesi  # Listeyi döndürelim.

# Pisagor üçlülerini listeden yazdıralım.
for i in pisagor_bulma():
    print(i)
```
